```python
import requests
import json
import pandas as pd
```

```python
test_runs = json.loads('[{"browser_name":"chrome","browser_version":"62.0","os_name":"linux","os_version":"3.16","revision":"10c9e8d229","results_url":"https://storage.googleapis.com/wptd/10c9e8d229/chrome-62.0-linux-summary.json.gz","created_at":"2017-10-04T14:21:11.38718Z"},{"browser_name":"edge","browser_version":"15","os_name":"windows","os_version":"10","revision":"fa273acc79","results_url":"https://storage.googleapis.com/wptd/fa273acc79/edge-15-windows-10-sauce-summary.json.gz","created_at":"2017-10-05T06:14:26.938244Z"},{"browser_name":"firefox","browser_version":"57.0","os_name":"linux","os_version":"3.16","revision":"85a171624e","results_url":"https://storage.googleapis.com/wptd/85a171624e/firefox-57.0-linux-summary.json.gz","created_at":"2017-10-03T18:46:24.067342Z"},{"browser_name":"safari","browser_version":"10","os_name":"macos","os_version":"10.12","revision":"3f00176743","results_url":"https://storage.googleapis.com/wptd/3f00176743/safari-10-macos-10.12-sauce-summary.json.gz","created_at":"2017-10-05T12:45:24.196748Z"}]')
url = test_runs[0].get('results_url')
res = requests.get(url)
data = res.json()
print(len(data), 'test files')
```

```python
rows = []
for filename, results in data.items():
    passing, total = results
    rows.append([filename, passing, total])
    
df = pd.DataFrame(data=rows, columns=['filename', 'passing', 'total'])
df.describe()
```
